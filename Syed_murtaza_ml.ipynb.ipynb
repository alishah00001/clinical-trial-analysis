```
from pyspark.sql.types import *
inputPath = "/FileStore/tables/steam_200k.csv"
myschema = StructType([
    StructField("member Id", DoubleType(), True),
    StructField("Name of the Game", StringType(), True),
    StructField("member behaviour", StringType(), True),  
    StructField("Status", DoubleType(), True),
])
# Read the data as a DataFrame with the specified schema
OccupancyDF = (
    spark.read
    .option("header", True)  
    .schema(myschema)
    .csv(inputPath)
)

display(OccupancyDF)
```

```
from pyspark.ml.feature import StringIndexer

# Create a StringIndexer to generate integer IDs for the game names
indexer = StringIndexer(inputCol="Name of the Game", outputCol="gameId")

# Fit the StringIndexer to the data and transform the DataFrame
indexed_df = indexer.fit(OccupancyDF).transform(OccupancyDF)

# Show the DataFrame with the generated integer IDs
indexed_df.show()
```

```
from pyspark.sql.functions import col, min as pyspark_min, max as pyspark_max
from pyspark.ml.feature import StringIndexer, MinMaxScaler
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator

# Calculate the minimum and maximum values of the "Status" column
min_status = indexed_df.select(pyspark_min("Status")).first()[0]
max_status = indexed_df.select(pyspark_max("Status")).first()[0]

# Normalize the "Status" column to a range between 0 and 1
indexed_df = indexed_df.withColumn("Normalized_Status", (col("Status") - min_status) / (max_status - min_status))
indexed_df.show(5)
```

```
# Split the data into training and test sets
(training, test) = indexed_df.randomSplit([0.8, 0.2], seed=100)
```

```
# Define the ALS model with the required parameters
als = ALS(maxIter=10, regParam=0.01, userCol="member Id", itemCol="gameId", ratingCol="Normalized_Status",
          coldStartStrategy="drop")
```

```
# Fit the ALS model to the training data
model = als.fit(training)
```

```
#  evaluate the model's performance on a test set
predictions = model.transform(test)
evaluator = RegressionEvaluator(metricName="rmse", labelCol="Normalized_Status", predictionCol="prediction")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data after normalization = {:.2f}".format(rmse))

# Generate predictions using the trained model
predictions.show()
```

```
from pyspark.ml.tuning import ParamGridBuilder

# Define the values for each parameter
maxIter_values = [5, 10, 15]
rank_values = [5, 10, 15]
regParam_values = [0.01, 0.1, 0.2]

# Create a parameter grid builder
paramGrid = (ParamGridBuilder()
             .addGrid(als.maxIter, maxIter_values)
             .addGrid(als.rank, rank_values)
             .addGrid(als.regParam, regParam_values)
             .build())
```

```
from pyspark.ml.tuning import TrainValidationSplit

# Instantiate TrainValidationSplit
tvs = TrainValidationSplit(estimator=als,
                            estimatorParamMaps=paramGrid,
                            evaluator=evaluator,
                            # 80% of the data will be used for training, 20% for validation
                            trainRatio=0.8)
```

```
# Train the model using grid search
model =tvs.fit(training)
```

```

# Get the best model from TrainValidationSplit using its getter method
best_model = model.getEstimator()


print("Parameters for Best Model:")


print("maxIter:", best_model.getMaxIter())
print("rank:", best_model.getRank())
print("regParam:", best_model.getRegParam())
print("RMSE:", rmse)
```

```
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit
from pyspark.ml.evaluation import RegressionEvaluator
# Define the values for each parameter
maxIter_values = [5, 10, 20]
rank_values = [4, 10, 18]
regParam_values = [0.01, 0.001, 0.2]
# Create a parameter grid builder
paramGrid = ParamGridBuilder() \
    .addGrid(als.maxIter, maxIter_values) \
    .addGrid(als.rank, rank_values) \
    .addGrid(als.regParam, regParam_values) \
    .build()
```

```

# Define the evaluator
evaluator = RegressionEvaluator(metricName="rmse", labelCol="Normalized_Status", predictionCol="prediction")
# Create TrainValidationSplit
tvs = TrainValidationSplit(estimator=als,
                            estimatorParamMaps=paramGrid,
                            evaluator=evaluator,
                            trainRatio=0.8)
# Fit TrainValidationSplit to the training data
model = tvs.fit(training)
# Get the best model from TrainValidationSplit
best_model = model.bestModel
# Make predictions on the test data
predictions = best_model.transform(test)
# Evaluate the predictions using the evaluation metric
rmse = evaluator.evaluate(predictions)
print("Best Model RMSE:", rmse)
```

```
# Print the hyperparameters and RMSE
print("Parameters for Best Model:")
print("maxIter:", best_model._java_obj.getMaxIter())
print("rank:", best_model._java_obj.getRank())
print("regParam:", best_model._java_obj.getRegParam())
print("Best Model RMSE:", rmse)
```
